### Connecting to Akida Cloud from Jupyter Lab

#### 1) From a Terminal, SSH to `akida-cloud`



#### 2) cd to `bc_cloud_examples` and run `start-jupyterlab.sh`



#### 3) Click on URL to launch Jupyter in a Browser.

(note: port `8888` is the tunnel between the local PC and the Akida Cloud server)
